```python
import pandas as pd
url = '/content/crabs.csv'
data = pd.read_csv(url)
import matplotlib.pyplot as plt 
data.hist()
plt.show
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns 
sns.heatmap(correlation,annot=True)
```

```python
data.plot(kind ='scatter',x='width',y='index',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['width']),y=data['index'])
print(model.intercept_)
print(model.coef_)
model.score(X=pd.DataFrame(data['width']),y=data['index'])
```

```python
data.plot(kind ='scatter',x='width',y='index',figsize=(9,9),color='black')
y_pred = model.predict(X=pd.DataFrame(data['width']))
plt.plot(data['width'],y_pred,color='green')
```

```python
model = linear_model.LinearRegression()
X=pd.DataFrame([data['width'],data['width']**2]).T
model.fit(X,y=data['index'])
print('Intercept cua mo hinh: ', model.intercept_)
print('Coeffient cua mo hinh: ', model.coef_)
print('Do chinh xac cua mo hinh: ', model.score(X,y=data['index']))
```

```python
import numpy as np
range = np.arange(1,55,0.1)
data.plot(kind ='scatter',x='width',y='index',figsize=(9,9),color='black')
predictor = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X=predictor)
plt.plot(range,y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X=pd.DataFrame([data['width'],data['width']**2]).T
model.fit(X,y=data['index'])
print('Intercept cua mo hinh: ', model.intercept_)
print('Coeffient cua mo hinh: ', model.coef_)
print('Do chinh xac cua mo hinh: ', model.score(X,y=data['index']))
```
